```python
from src.data_loader import load_posts
from src.recommender import HashtagRecommender
from src.preprocess import clean_text

df = load_posts('../data/raw/social_posts_synthetic.csv')
model = HashtagRecommender().fit(df)

query = 'ranked grind tonight'
pred = model.recommend(clean_text(query), top_k=5, top_tags=6)
print('Query:', query)
print('Suggested hashtags:', ' '.join(pred.hashtags))
print('Closest captions:')
for cap, score in pred.neighbors:
    print(f'{score:.3f}', cap)
```
